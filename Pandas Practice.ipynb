```python
# use default setting to show dataframe
df = pd.read_csv('http://bit.ly/kaggletrain')
display(df.head(5))
```

```python
# Define the max column width of "Name" column
from IPython.display import display

pd.set_option("display.max_colwidth", 10)
df.head(5)
```

```python
# Modify float for example, columns named "Fare" has changed
pd.set_option("display.precision", 1)
df.head(5)
```

```python
pd.reset_option("all")
```

```python
# randomly choose 20 rows for styling
df_sample = df.sample(n=10, random_state=9527).drop('Name', axis=1)
df_sample.Age.fillna(int(df.Age.mean()), inplace=True)
df_sample
```

```python
(df_sample.style
     .format('{:.1f}', subset='Fare')
     .set_caption('colorful dataframe')
     .hide_index()
     .bar('Age', vmin=0)
     .highlight_max('Survived')
     .background_gradient('Greens', 
                          subset='Fare')
     .highlight_null()
)
```
